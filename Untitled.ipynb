# KMeansClassifier
Здесь создается класс KMeansClassifier. Конструктор ринимает 2 параметра:<br>
1) эпсилон - точность вычисления. В данном случае эпсилон характеризует количество пикселей, на которое сместился центр кластера. Вычисления прекращаются, как только каждый центр сместился на <= eps пикселей<br>
2) mode: если 0, то выбор начальных центров производится случайно; если 1, то используется k-means++<br>
<hr>
Выбор центров в режиме 1 происходит следующим образом:<br>
1) первый центр выбирается случайно<br>
2) для каждой точки вычисляется квадрат расстояния до ближайшего центра из уже имеющихся<br>
3) выбираем новый центр. Основная идея: вероятность точки стать новым центром пропорциональна квадрату расстояния от нее до ближайшего центра.<br>
Поступим следующим образом. Посчитаем сумму всех квадратов расстояний (sum) и сгенериурем случайное число (rnd) в диапазоне [0; sum). При этом значения квадратов расстояний для каждой точки мы запомнили на предыдущем шаге. Пройдемся по массиву квадратов расстояний и начнем суммировать сначала. Как только сумма превысит rnd, берем точку за новый центр.<br>
Сам алгоритм заключается в том, что мы в цикле сначала относим каждую точку к какому-либо из кластеров, а затем перевычисляем центры кластеров (усредняя все точки из этого кластера). КРитерий останова описан выше.

```python
class KMeansClassifier(object):
    """docstring"""
    
    def __init__(self, eps = 0.006, mode = 1): #тут эпсилон - это насколько сдвинулись центры кластеров
        """Constructor"""
        self.stop = False
        self.eps = eps
        self.mode = mode
    
    class Image(object):
        """docstring"""

        def __init__(self, data):
            """Constructor"""
            self.data = data
            self.cluster = -1

    
    def distance(self, p1, p2):
        return la.norm(p1 - p2)
    
    def divide_data(self):
        index = 0
        for i in xrange(0, len(self.clusters)):
            self.clusters[i] = []
            
        for i in xrange(0, len(self.x)):
            min_dist = None
            image = self.x[i]
            for j in xrange(0, len(self.centers)):
                cur_dist = self.distance(image.data, self.centers[j])
                if (min_dist is None) | (cur_dist < min_dist):
                    min_dist = cur_dist
                    index = j
            if image.cluster != index:
                image.cluster = index
            self.clusters[index].append(image)
    
    
    def refind_centers(self):
        prev_centers = []
        size = len(self.clusters[0][0].data)
        for i in xrange(0, len(self.clusters)):
            sum = np.zeros(size)
            for j in xrange(0, len(self.clusters[i])):
                sum = sum + self.clusters[i][j].data
            prev_centers.append(self.centers[i])
            self.centers[i] = sum / len(self.clusters[i])

        self.stop = True
        for i in xrange(0, len(self.centers)):
            if la.norm(prev_centers[i] - self.centers[i]) * 1. / size >= self.eps:
                self.stop = False
                break

    def choose_centers(self, mode):
        self.centers = []
        
        if mode == 0:
            tmp = np.random.choice(self.x, 10)
            for i in xrange(0, len(tmp)):
                self.centers.append(tmp[i].data)
            return
        
        #mode = 1 -> kmeans++
        rand_index = np.random.randint(0, len(self.x))
        self.centers.append(self.x[rand_index].data) #первую точку выбираем случайно
        
        num = 0
        while (len(self.centers) < 10):
            sum = 0
            dx2 = []
            for i in xrange(0, len(self.x)):
                min_sq = None
                for j in xrange(0, len(self.centers)):
                    cur_dist = self.distance(self.centers[j], self.x[i].data)
                    if cur_dist == 0:
                        num += 1
                    if (min_sq is None) | (min_sq > cur_dist * cur_dist):
                        min_sq = cur_dist * cur_dist                    
                sum += min_sq
                dx2.append(min_sq)
            
            rnd = np.random.random() * sum
            s = 0
            for i in xrange(0, len(dx2)):
                s += dx2[i]
                if s > rnd:
                    self.centers.append(self.x[i].data)
                    break
                        
    def classify(self, x):     
        self.x = []
        for i in xrange(0, len(x)):
            self.x.append(self.Image(x[i]))
            
        self.choose_centers(self.mode)
            
        self.clusters = []
        for i in xrange(0, len(self.centers)):
            self.clusters.append([])
        self.divide_data()
        self.iter = 1
        while True:
            self.refind_centers()
            self.divide_data()
            self.iter += 1
            if self.stop == True:
                break
            
            
    def print_digits(self, num):
        for i in xrange(0, len(self.clusters)):
            print i
            for j in xrange(0, num):
                print_digit(np.array(self.clusters[i][j].data).reshape(28, 28))
                
    def check(self, adapt, answer):
        right = 0
        for i in xrange(0, len(self.x)):
            if adapt[self.x[i].cluster] == answer[i]:
                right += 1
        print ('accuracy', 1. * right / len(self.x))
```

# Построение диаграммы Вороного
Построим диаграмму Вороного за O(n<sup>4</sup>). Кратко алгоритм:<br>
1) зафиксировать один центроид p<br>
2) провести отрезки, соединяющий p и остальные центроиды<br>
3) построить к отрезкам из п. 2 серединные перпендикуляры<br>
4) пересечь серединные перпендикуляры между собой<br>
5) пересечь серединные перпендикуляры с bounding box<br>
6) для каждой точки пересечения посмотреть, лежит ли она в той же полуплоскости, что и p, относительно всех серединных перпендикуляров: если да, то она будет являться одной из вершин многоугольника, ограничивающего область влияния p<br>
7) повторить пункты 1 - 6 для всех центроидов<br>
8) определить, к области влияния какого из центроидов относятся углы bounding box<br>
9) теперь для каждого центроида есть множество точек, задающих многоугольник. Надо расположить их в нужном порядке (чтобы получился выпуклый). Используем метод Грэхэма построения минимальной выпуклой оболочки (в данном случае в нее войдут все точки)<br>
10) для каждого центроида рисуем соответствующий многоугольник<br>
# Анализ результата
Полученная картинка отличается от картинки, полученной в результате применения стандартной реализации алгоритма. Это объясняется тем, что положение центров немного смещено. Даже при небольшом смещении точек формы многоугольников могут значительно отличаться.

```python
def get_ser_per(p1, p2):
    x1 = p1[0]
    x2 = p2[0]
    y1 = p1[1]
    y2 = p2[1]
    a = 2 * x2 - 2 * x1
    b = 2 * y2 - 2 * y1
    c = x1 * x1 + y1 * y1 - x2 * x2 - y2 * y2
    koeff = [a, b, c]
    return koeff

def intersection_point(line1, line2):
    a1 = line1[0]
    b1 = line1[1]
    c1 = line1[2]
    a2 = line2[0]
    b2 = line2[1]
    c2 = line2[2]
    if b1 == 0:
        if b2 == 0:
            return None
        else:
            x0 = -c1 * 1. / a1
            y0 = (-c2 - a2 * x0) / b2
            return [x0, y0]
    if b2 == 0:
        #b1 != 0
        x0 = -c2 * 1. / a2
        y0 = (-c1 - a1 * x0) / b1
        return [x0, y0]
    #имеем 2 невертикальные прямые
    if a2 * b1 == a1 * b2:
        return None
    x0 = (c2 * b1 - c1 * b2) * 1. / (a1 * b2 - a2 * b1)
    y0 = (a1 * x0 + c1) * -1. / b1
    return [x0, y0]

def are_in_one_halfplane(line, p1, p2):
    #print (p1, p2)
    eps = 0.001
    t1 = line[0] * p1[0] + line[1] * p1[1] + line[2]
    t2 = line[0] * p2[0] + line[1] * p2[1] + line[2]
    if t1 * t2 >= -eps:
        return True
    return False

def compare(p1, p2, x0, y0):
    #print (p1, p2)
    if p1[0] == 0:
        if p2[0] == 0:
            d1 = (p1[0] - 0) ** 2 + (p1[1] - y0) ** 2
            d2 = (p2[0] - 0) ** 2 + (p2[1] - y0) ** 2
            if d1 > d2:
                return 1
            else:
                if p2[0] > 0:
                    return -1
                else:
                    return 1
        else:
            return -1
    if p2[0] == 0:
        if p1[0] > 0:
            return -1
        else:
            return 1
    tg1 = (p1[1] - y0) * 1. / (p1[0] - 0)
    tg2 = (p2[1] - y0) * 1. / (p2[0] - 0)
    
    if abs(tg1) > abs(tg2):
        return 1
    elif abs(tg1) < abs(tg2):
        return -1
    d1 = (p1[0] - 0) ** 2 + (p1[1] - y0) ** 2
    d2 = (p2[0] - 0) ** 2 + (p2[1] - y0) ** 2
    if d1 > d2:
        return -1
    else:
        return 1
    
def rotate(A, B, C):
    return (B[0] - A[0]) * (C[1] - B[1]) - (B[1] - A[1]) * (C[0] - B[0])
    
def mvo(points):
    n = len(points)
    P = range(n) 
    for i in xrange(1 ,n):
        if points[P[i]][0] < points[P[0]][0]:
            P[i], P[0] = P[0], P[i]
    for i in xrange(2, n):
        j = i
        while (j > 1) & (rotate(points[P[0]], points[P[j-1]], points[P[j]]) < 0): 
            P[j], P[j-1] = P[j-1], P[j]
            j -= 1
    S = [P[0],P[1]]
    for i in xrange(2, n):
        while rotate(points[S[-2]], points[S[-1]], points[P[i]])<0:
            del S[-1]
        S.append(P[i])
    return S
           
def contains_point(arr, p):
    eps = 0.0000001
    for i in xrange(0, len(arr)):
        if la.norm(np.array(arr[i]) - np.array(p)) <= eps:
            return True
    return False
    
def voronoi(points, x_min, x_max, y_min, y_max):
    vertices = []
    ser_pers = []
    for i in xrange(0, len(points)):
        vertices_p = []
        ser_per = []
        intersection_points = []
        for j in xrange(0, len(points)):
            if j == i:
                continue
            ser_per.append(get_ser_per(points[i], points[j]))
        for j in xrange(0, len(ser_per)):
            for k in xrange(0, len(ser_per)):
                if j == k:
                    continue                
                ip = intersection_point(ser_per[k], ser_per[j])
                if ip is not None:
                    if (ip[0] >= x_min) & (ip[0] <= x_max) & (ip[1] >= y_min) & (ip[1] <= y_max):
                        intersection_points.append(ip)
            ip = intersection_point(ser_per[j], [1, 0, -x_min])
            if ip is not None:
                if (ip[0] >= x_min) & (ip[0] <= x_max) & (ip[1] >= y_min) & (ip[1] <= y_max):
                    intersection_points.append(ip)
            ip = intersection_point(ser_per[j], [1, 0, -x_max])
            if ip is not None:
                if (ip[0] >= x_min) & (ip[0] <= x_max) & (ip[1] >= y_min) & (ip[1] <= y_max):
                    intersection_points.append(ip)
            ip = intersection_point(ser_per[j], [0, 1, -y_min])
            if ip is not None:
                if (ip[0] >= x_min) & (ip[0] <= x_max) & (ip[1] >= y_min) & (ip[1] <= y_max):
                    intersection_points.append(ip)
            ip = intersection_point(ser_per[j], [0, 1, -y_max])
            if ip is not None:
                if (ip[0] >= x_min) & (ip[0] <= x_max) & (ip[1] >= y_min) & (ip[1] <= y_max):
                    intersection_points.append(ip)
            
        cur = np.array(intersection_points).reshape(-1, 2)
            
        ips = np.array(intersection_points).reshape(-1, 2)
        for j in xrange(0, len(intersection_points)):
            if_vert = True
            for k in xrange(0, len(ser_per)):
                if are_in_one_halfplane(ser_per[k], intersection_points[j], points[i]) == False:
                    if_vert = False
                    break
            if if_vert & (contains_point(vertices_p, intersection_points[j]) == False):    
                vertices_p.append(intersection_points[j])  
        vertices.append(vertices_p)
        ser_pers.append(ser_per)

    angles = [[x_min, y_min], [x_min, y_max], [x_max, y_min], [x_max, y_max]]
    for i in xrange(0, len(angles)):
        cur_min = None
        index = 0
        for j in xrange(0, len(points)):
            cur_dist = la.norm(np.array(points[j]) - np.array(angles[i]))
            if (cur_min is None) | (cur_dist < cur_min):
                cur_min = cur_dist
                index = j
        if contains_point(vertices[index], angles[i]) == False:
            vertices[index].append(angles[i])
        
    p = np.array(points).reshape(-1, 2)
    for i in xrange(0, len(vertices)):
        cur = []
        seq = mvo(vertices[i])
        for j in xrange(0, len(seq)):
            cur.append(vertices[i][seq[j]])
        cur = np.array(cur).reshape(-1, 2)
        color = [np.random.rand(), np.random.rand(), np.random.rand()]
        plt.fill(cur[:, 0], cur[:, 1], c = color, alpha = 0.3)
    plt.scatter(p[:, 0], p[:, 1], c = 'r')
    plt.show(label = 'Voronoi diagram')
```
